Задание 1 Переведите содержимое файла purchase_log.txt*
в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, …}

```python
import json
dic={}
list_=[]
f=open('purchase_log.txt','r')
n=7
for line in f:
    dict_ = json.loads(line)
    key_=dict_['user_id']
    value_=dict_['category']
    dic[key_]=value_

    if n>0:
        print(dic)
        n=n-1
f.close()
```

Задание 2 Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки, если покупка была, сам файл visit_log.csv* изменять не надо. Запишите в файл funnel.csv визиты из файла visit_log.csv*, в которых были покупки с указанием категории.
Учтите условия на данные:
содержимое purchase_log.txt* помещается в оперативную память компьютера;
содержимое visit_log.csv* — нет; используйте только построчную обработку этого файла.

```python
user_id=''
f=open('visit_log.csv','r')
ff=open('funnel.csv','a')
for line in f:
    list_=line.strip().split(',')
    user_id=list_[0]
    if user_id in dic:
        ff.write(list_[0]+','+list_[1]+','+dic[user_id]+'\n')
f.close()
ff.close
fff=open('funnel.csv','r')
for line in fff:
    print(line)
fff.close()
```
